```python
def solution(n):
    product = 1
    has_odd_digit = False
    
    while n > 0:
        digit = n % 10
        if digit % 2 == 1:  # if digit is odd
            product *= digit
            has_odd_digit = True
        n = n // 10
    
    return product if has_odd_digit else 0

print(solution(1234))  # Example usage
```
